### Visualizing Distribution of Dataset

Plotting univariate distributions

We can plot univariate distribution in seaborn using the distplot() function.The example below explains this:

```python
#Example of univariate distribution plot
x = np.random.normal(size=100)
sns.distplot(x);
```

Plotting bivariate distributions

We can plot bivariate distributions using the jointplot() function, which creates a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
sns.jointplot(x="x", y="y", data=df);
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```

### Visualizing relationships of Data

Pairplot:
    To plot pairwise relationship in data in seaborn , pairplot() function is used as shown below in example for iris dataset.

```python
#Visualizing pairwise relationship in data
iris = sns.load_dataset("iris")
sns.pairplot(iris);
```

Visualizing Linear relationships:
    Seaborn provides regplot(),lmplot() functions to visualize the linear relationships in data.The following examples illustrates this.

```python
#Showing linear relationship
tips = sns.load_dataset("tips")
sns.regplot(x="total_bill", y="tip", data=tips);
```

Fitting model:

    Using seaborn we can fit different kinds of models like linear , quadratic for the given data using lmplot() specifying the order as shown in the examples below:
    

```python
#linear model
anscombe = sns.load_dataset("anscombe")
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
           ci=None, scatter_kws={"s": 80});
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           ci=None, scatter_kws={"s": 80});
```

```python
#Fitting non linear model
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80});
```

### SCATTER PLOT

Scatter plot of data can be made using relplot(), scatterplot(), catplot() functions as shown in example below:

```python
#Scatter Plot

#single value scatter plot
mpg = sns.load_dataset("mpg")
print(mpg.columns)
#sns.relplot(x="horsepower", y="acceleration", data=mpg);
sns.scatterplot(x="horsepower",y="acceleration",data=mpg)

#multiple scatter plot
tips= sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);

#categorical scatter plot
tips = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=tips);

#try this
#sns.catplot(x="day", y="total_bill", kind="swarm", data=tips);swarm plot 


```

## BOX PLOTS
Box plot of data can be made using catplot() function as shown in example below:

```python
#Load data
data = sns.load_dataset("tips")
print(data.columns)

#simple box plot
sns.catplot(x="day", y="total_bill", kind="box", data=tips);


#box plot with hue

sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips);

```

### BAR CHARTS
Bar plot of data can be made using catplot(),barplot() functions as shown in example below:

```python
#Loading the dataset
titanic = sns.load_dataset("titanic")
print("Titanic dataset features:",titanic.columns)
tips =sns.load_dataset("tips")
print("Tips dataset features:",tips.columns)
#Simple Bar chart
plt.figure() #Used to remove overlapping of plots 
sns.catplot(x="deck", kind="count", palette="ch:.25", data=titanic);

#Multiple bar chart
plt.figure()
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic);

#count bar plot
plt.figure()
sns.countplot(y="deck", data=titanic, color="c");

#bargraph using barplot
plt.figure()
ax = sns.barplot(x="tip", y="day", data=tips)
```
