### Parts of figure from matplotlib

#### Basic concepts:
- Title:
    It is the heading of the plot that appears at the top of the figure.
- Axes:
     It is the region of the image with the data space.This is shown in the figure below
- Axis:
    These are the number-line-like objects. They take care of setting the graph limits and generating the ticks (the marks on the axis) and ticklabels (strings labeling the ticks).For more clearance see figure below
    
- Grid:
    It is the dotted lines that divides the axes into multiple equal square spaces.

- label:
    It is the notation for axis.For instance we can define x-axis label and y-axis label names. 
    
- Legend:
    It is the notation for plots in the figure.It is often represented with distinct colors ,shapes and names.
    
- Figure:
    A figure in matplotlib means the whole window in the user interface. Within this figure there can be “subplots” that helps to arrange plots in a regular grid where need to specify the number of rows and columns and the number of the plot

Subplots:
Matplotlib allows to arrange multiple data plots in rectangular grid using subplot() function where we need to specify the number of rows and columns and the plot number as shown in example below:
    

```python
#Example of subplot

#Format: subplot(x,y,z); x=Total no.of rows in subplot grid,y=Total no.of columns in subplot grid, z=position/number for particular plot

plt.figure()

plt.subplot(2, 2, 1)

#Writing text inside subplot
plt.text(0.5, 0.5, 'subplot(2,2,1)', ha='center', va='center',
        size=20, alpha=.5)

plt.subplot(2, 2, 2)

#Writing text inside subplot
plt.text(0.5, 0.5, 'subplot(2,2,2)', ha='center', va='center',
        size=20, alpha=.5)

plt.subplot(2, 2, 3)

#Writing text inside subplot
plt.text(0.5, 0.5, 'subplot(2,2,3)', ha='center', va='center',
        size=20, alpha=.5)

plt.subplot(2, 2, 4)

#Writing text inside subplot
plt.text(0.5, 0.5, 'subplot(2,2,4)', ha='center', va='center',
        size=20, alpha=.5)

plt.show()
```

```python
#Example of plotting with default settings

X = np.linspace(0, 50, 100, endpoint=True) #numpy array to specific ranges
E = np.exp(X) #using numpy exponential function
plt.plot(X, E) #Here we are using plot with default settings
plt.show()
```

```python
# Changing the title,labels,colors,linewidth,limit and figsize

#Preparing data
X = np.linspace(0, 50, 100, endpoint=True)
E = np.exp(X) 

#defining figure
fig=plt.figure(figsize=(6, 4), dpi=80)

#Making a plot
plt.plot(X,E,color="green",linewidth=2.5) 

# Add title and axis names
fig.suptitle('CustomizedFig(i)', fontsize=10)
plt.title('Exponential Graph')
plt.xlabel('X', fontsize=10)
plt.ylabel('E', fontsize=10)

#Show the plot
plt.show()
```

```python
#Setting ticks

#Preparing data
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S = np.sin(X)

#Make plot
plt.plot(X,S)

#Define xticks and yticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi]) #defining x ticks
plt.yticks([-1, 0, +1])                        #defining y ticks

#Show the plot
plt.show()
```

```python
#Setting tick labels

#Prepare data
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S = np.sin(X)

#Make plot
plt.plot(X,S)

#Set ticks label
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

#Show the plot
plt.show()
```

```python
# Adding a legend
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S = np.sin(X)
C = np.cos(X) 
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine") #declaring label name of the plot that will be used in legend
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label="sine")
plt.legend(loc='upper left')
plt.show()
```

```python
#Annotating text in plot

#Prepare data and make plot
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line= plt.plot(t, s)

#annotate in particular point in plot
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

#Set y value limit
plt.ylim(-2, 2)

#Show the result
plt.show()
```

## LINE PLOT
In matplotlib,line plots can be made using plot() function specifying the x-axis and y-axis data as shown in example below:

```python
# Simple line plot 

#Prepare data
year=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
total_items_sales=[2555,3666,4445,6666,7586,9523,8658,12563,13256,15256]

#MAke the line plot 
plt.plot(year,total_items_sales )
plt.ylabel('Total no of items sales')
plt.xlabel('Year')

#show the plot
plt.show()

#Make and show the point plot with red circle 
plt.plot(year,total_items_sales,'ro')
plt.show()

#Make and show the point plot with blue square 
plt.plot(year,total_items_sales,'bs')
plt.show()


```

### SCATTER PLOT
Scatter plot of data can be made using scatter() function by specifying plot data as shown in example below:

```python
# Basic scatter plot

#Preparing data by creating some random data dictionary 
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

#Making a scatter plot
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel(' a')
plt.ylabel(' b')
plt.show()
```

### BAR PLOT
Bar plots are made using bar() and barh() functions as shown in example below:

```python
# plotting with categorical values

#Preparing data
cake_items = ['kitkat', 'Unicorn', 'Chocolateroll', 'Barbiedoll','Doraemon']
no_cake_items_sales_month=[235,554,582,695,545]

#plotting figure
plt.figure(figsize=(15, 5))

#defining subplots and making respective plots
plt.subplot(221)
plt.bar(cake_items,no_cake_items_sales_month)
plt.subplot(222)
plt.scatter(cake_items,no_cake_items_sales_month)
plt.subplot(223)
plt.plot(cake_items,no_cake_items_sales_month)
plt.subplot(224)
plt.barh(cake_items,no_cake_items_sales_month,color="green")
plt.suptitle('Different Categorical Plottings')
plt.show()
```

```python
# Stacked bar graph

#Preparing data
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

#Make two different bar plots
p1 = plt.bar(ind, menMeans, width, yerr=menStd)
p2 = plt.bar(ind, womenMeans, width,
             bottom=menMeans, yerr=womenStd)

#Set the ylabel and title
plt.ylabel('Scores')
plt.title('Scores by group and gender')

#Set xticks and yticks
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))

#Set the legend 
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

#Show the final plot
plt.show()
```

### PIE CHART

Pie charts can be made using pie() function as shown in the example below:


```python
#simple pie chart

#Preparing data
cake_items = ['kitkat','Barbiedoll', 'Unicorn', 'Chocolateroll','Doraemon']
no_cake_items_sales_month=[235,695,554,550,545]

#Defining pie chart elements
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
Explode = (0, 0.1, 0, 0,0)  # only "explode" the 2nd slice (i.e. 'Barbiedoll')

#Make the suitable pie plot and show the plot
plt.pie(no_cake_items_sales_month,labels=cake_items,explode=Explode,startangle=45)
plt.axis('equal')
plt.legend(title="Cake Items Sales", bbox_to_anchor=(1.2,0), loc="lower right",bbox_transform=plt.gcf().transFigure)# bbox_to_anchor
plt.show()
```

### 3D PLOTS

 3D graphs can be plotted using Matplotlib's toolkit mpl_toolkits. Using this tool 3Dline, scatter3D ,surface,contour,polygons and much more other 3D plots can be made.Few simple examples of these plots are shown below:



```python
#Simple 3D space plot 

#Import toolkit for 3d
from mpl_toolkits import mplot3d
%matplotlib inline 

#Define figures and axes
fig = plt.figure(figsize=(8, 4), dpi=80)
ax = plt.axes(projection='3d')

#Show the result
plt.show()
```

```python
#Simple Example of line and points plot in 3D space

fig = plt.figure(figsize=(8, 4), dpi=80)
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'blue')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)

#Make scatter 3D plot and label axis
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='hot');
ax.set_xlabel('X axis',fontsize=8)
ax.set_ylabel('Y axis',fontsize=8)
ax.set_zlabel('Z axis',fontsize=8)

#Show the result
plt.show()
```

### Working with Images

To work with images first import the image data into Numpy arrays using mpimg()function . Then plot this array in the figure and customize the plot.


```python
#Import the image tool from matplotlib
import matplotlib.image as mpimg

#Read the image and store as numpy array
img = mpimg.imread('https://upload.wikimedia.org/wikipedia/commons/4/47/PNG_transparency_demonstration_1.png')
print(type(img))
#print(img) #prints numpy arrays of numbers 
```

```python
imgplot = plt.imshow(img)
```

```python
#Applying pseudocolor schemes to image plots(for enhancing contrast and visualizing data more easily)
lum_img = img[:, :, 0]
plt.imshow(lum_img)
```
