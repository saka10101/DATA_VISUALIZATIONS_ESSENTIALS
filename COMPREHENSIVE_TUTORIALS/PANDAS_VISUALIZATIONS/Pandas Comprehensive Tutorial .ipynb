##### SERIES


Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.) 
For example to create a simple series:

s = pd.Series(data, index=index)

here ,data can be many different things like:Python dictionary,nd array,scalar value(like :10) and we can also specify the index of data from argument( axis labels are collectively referred to as the index).
There are several ways of creating series which are explained with examples below:


```python

#Example of creating pandas series using some random values list
s = pd.Series([1, 8, 5, np.nan, 6, 7],name="defined series") #Note:NaN (not a number) is the standard missing data marker used in pandas.
print(s) #prints out the series
print(s.index) #prints index of Series



```

```python
#Example of creating pandas series using dictionary
dictionary={'a':1,'b':2,'c':3}
s_dict=pd.Series(dictionary)
print(s_dict)
print(s_dict.index)
#Get value of series like dictionary
print("Value of b is:",s_dict['b'])#prints out value in b
```

```python
#Slicing operations using series
s = pd.Series([1, 8, 5, 25, 6, 7],name="new series")
#General format for slicing s[start:end:step] like list slicing
print(s[:2]) #prints first two rows
print(s[1:4:1])#prints from 8 to 6
```

Operations on Series:
Pandas series supports variety of mathematical operations.Some of them is shown in examples below:

```python
#Operations on Series

s = pd.Series([1, 8, 5, 25, 6, 7],name="new series")
print("s: \n",s)
#Vectorized operations in series
print("\ns+s:\n",s+s) #just adds the values like vector
print("\ns-s:\n",s-s) #just subtracts the values 
print("\ns*2:\n",s*2) #just multiples by 2
print("\ns/2:\n",s/2) #just divides by 2
print("\ne^s:\n",np.exp(s))  #finds exponent
print("\nlog(s):\n",np.log(s)) #finds log
```

#### DATAFRAME

DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It is like a spreadsheet or SQL table, or a dict of Series objects.
For example to create a simple dataframe:

df = pd.DataFrame(data=data,index=index)

DataFrame accepts many different kinds of data inputs like Dict of 1D ndarrays, lists, dicts, or Series,2-D numpy.ndarray,structured or record ndarray,another DataFrame.Along with the data, you can optionally pass index (row labels) and columns (column labels).Different methods for creating dataframes are explored in the examples below:

```python
#Creating dataframe from dictionary of nd arrays or lists
d = {'one': [1, 2, 3, 4],
    'two': [4, 3, 2, 1]}
df0=pd.DataFrame(d)
print(df0)
print(df0.index)
```

```python
#Creating dataframe from dictionary of series 
dict_data= {'Column_one': pd.Series([1, 2, 3], index=['a', 'b', 'c']),
     'Column_two': pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])}
df1=pd.DataFrame(dict_data)
print(df1)
df1.index
```

```python

#Creating dataframe from structured or record array
data = np.zeros((2, ), dtype=[('A', 'i4'), ('B', 'f4'), ('C', 'a10')])
data[:] = [(1, 2., 'Hello'), (2, 3., "World")]
df2=pd.DataFrame(data)
print(df2)
```

```python
#Creating random valued dataframe using random 
df_1= pd.DataFrame(np.random.randn(5, 4), columns=['A', 'B', 'C', 'D'])
print(df_1)
```

```python
#Inspecting the dataframe
d = {'one': [1, 2, 3, 4],
    'two': [4, 3, 2, 1],
     'three':[3,8,2,8] }
df=pd.DataFrame(d)
```

Addition of columns in dataframe:
New columns can be easily added to dataframe in the same way data is added to the python dictionary.Take a look at examples below:

```python
#Addition of new column
df0=pd.DataFrame({'one': [1, 2, 3, 4],
                'two': [4, 3, 2, 1]})
df0['three'] = df0['one'] * df0['two'] #New column 'three' added by multiplying column 'one' and 'two' 
print(df0)
df0['flag'] = 1 #adds col 'flag' and fills down value 1
print(df0)
```

#### LOADING DATA FROM EXTERNAL SOURCES INTO DATAFRAMES

#I/O 

Pandas contains set of data reader functions like read_csv,read_json,read_excel that generally return a pandas object ie Dataframes or Series. 
Also the corresponding writer functions like to_csv,to_json,to_excel are object methods that helps convert pandas object into specified format. Pandas works with variety of file for loading data. Some I/O operations



Handling CSV & text files

CSV & text files(flat files) can be easily loaded into pandas using reader read_csv().
 Format:
 data=pd.read_csv("path_of_file_stored")
 eg:data.read_csv("../test.csv")
 where, 
 - path can be local file directory or web link to file 
 - data is the dataframe where data from file is loaded
 - other optional parameters can be specified such as index_col,n_rows,header,skiprows
 
To convert the dataframe to csv:
  data.to_csv("path_to_file_storage")
  eg:data.to_csv("../test2.csv")


```python
#Example of handling CSV & text files

#load data into dataframe
data=pd.read_csv("https://github.com/plotly/datasets/raw/master/Emissions%20Data.csv",index_col=None) #you can specify index column
data.head(4)

#Export Dataframe to csv
#data.to_csv('Path where you want to store the exported CSV file\File Name.csv')



```

Handling JSON files

JSON files can be imported using reader read_json()
Format:
data = pd.read_json ('Path where the json file is stored\File name.json')

Conversely you can also convert the pandas dataframe to json file using to_json()
data.to_json ('Path where you want json file to be stored\File name.json')



Handling excel files 

You can easily import an Excel file in pandas using reader read_excel()
Format:
data = pd.read_excel ('Path where Excel file stored\File name.xlsx', sheet_name='your Excel sheet name')

Conversely you can also convert the pandas dataframe to excel file using to_excel()
data.to_excel ('Path where you want excel file to be stored\File name.xlsx')

##### LINE PLOT
Line plot of data can be made using plot() function of dataframe object as shown in example below:

```python
#Simple line plot
data=pd.read_csv("https://github.com/plotly/datasets/raw/master/2014_apple_stock.csv",index_col=0)
data.plot()
plt.xlabel('Time')
plt.ylabel('value')

```

##### SCATTER PLOT
Scatter plot of data can be made using plot.scatter() function of dataframe object as shown in example below:

```python
#Simple Scatter Plot Example
df = pd.DataFrame([[5.1, 3.5, 0], [4.9, 3.0, 0], [7.0, 3.2, 1],
                   [6.4, 3.2, 1], [5.9, 3.0, 2]],
                  columns=['length', 'width', 'species'])
ax1 = df.plot.scatter(x='length',
                      y='width',
                      c='DarkBlue')
```

##### HISTOGRAM PLOT
Histogram plot of data can be made using plot.hist() function of dataframe object as shown in example below:

```python
#Simple histogram Plot
'''
When we draw a dice 6000 times, we expect to get each value around 1000 times.
But when we draw two dices and sum the result, the distribution is going to be quite different.
The histogram below shows this
'''

df = pd.DataFrame(
    np.random.randint(1, 7, 6000),
    columns = ['one'])
df['two'] = df['one'] + np.random.randint(1, 7, 6000)
ax = df.plot.hist(bins=12, alpha=0.5)

```

##### BAR GRAPH
Bar plot of data can be made using plot.bar() function of dataframe object as shown in example below:

```python
#Simple Bar chart

df = pd.DataFrame({'lab':['A', 'B', 'C'], 'val':[10, 30, 20]})
ax = df.plot.bar(x='lab', y='val', rot=0)

```
